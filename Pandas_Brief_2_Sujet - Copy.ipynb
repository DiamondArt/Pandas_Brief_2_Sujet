```python
#Lister les 15 produits les moins chers

df.sort_values(by=['Prix'])[:15]
```

```python
#Calculer la moyenne de prix des boissons distribuées à Lyon (pivot_table)
tb=pd.pivot_table(data=df,index=['Ville','Categorie'], values='Prix')
tb.head(1)
```

```python
#Quels sont les 5 produits les moins chers vendus à Lyon ?
df.sort_values(by=['Prix']).where(df['Ville'] == 'Lyon')[:5]
# df.sort_values(by=['Prix'])[:15]
```

```python
#En utilisant une boucle afficher la representation graphique de toutes les variables quantitative
qt= df.select_dtypes(exclude='object').columns
plt.figure(figsize=(35,5))

for index,item in enumerate(qt):
    i= index+1
    plt.subplot(1,7,i)
    df[item].value_counts().plot(kind='hist', edgecolor='white')
    i+=1
    plt.title(item)
```

```python
#En utilisant une boucle afficher la representation graphique de toutes les variables quanlitative
ql= df.select_dtypes(include='object').columns

for index,item in enumerate(ql):
    if item == 'Nom':
        plt.figure(figsize=(35,5))
    else:
        plt.figure(figsize=(10,5))   
    df[item].value_counts().plot(kind='bar')
    plt.title(item)
```
